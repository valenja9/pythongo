#Úkol 8: Rozšifruj tuto zprávu. Najdeš zde nové jméno manžela Leslie Jasona. Zkus měnit klíč naší šifry tak, aby ti vyšlo dobré jméno! Nezapomeň, že musíš prověřit všechny klíče od 1-300.
---
- Jsi přeci programátor/ka, a nebudeš jako cvičená opice zkoušet všechny čísla od 1 do 300. Zkus použít ještě jeden FORCYKLUS a nezapomeň, že při dešifrování je třeba změnit znaménko. Pak už bude jednoduché najít ten jediný smysluplný výsledek mezi těmi nesmysly.

```
MESSAGE = "ppfcbodtxfgc "
TABLE = 'abcdefghijklmnopqrstuvwxyz '
KEY = 1 # Klíč je třeba měnit

dict_ = {char:i for char,i in zip(TABLE,range(0,27))}
redict = {i:char for char, i in zip(TABLE,range(0,27))}

nova_zprava = ""
for char in MESSAGE:
    nova_zprava+=char.replace(char,redict[((dict_[char]+KEY)%len(TABLE))]) # ... a co to znaménko, nezapomněl/a jsi?


#Sem napiš forcyklus a zkus najít správné řešení

nova_zprava
```
