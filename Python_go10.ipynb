```
#@title
import requests
import os
from bs4 import BeautifulSoup
from urllib.parse import urljoin



def odkazy(odkaz):
    r = requests.get(odkaz)
    html = r.text

    soup = BeautifulSoup(html, 'html.parser')

    href_list = []
    for a_elem in soup.select("h2.css-1kiihh0 a"):
        print(a_elem["href"])
        href = urljoin("https://www.vzdelanibudoucnosti.cz/", a_elem["href"])
        print(50*"-")
        href_list.append(href)
    return  href_list

def nadpis(adresa_clanku):

    r = requests.get(adresa_clanku)
    html = r.text

    soup = BeautifulSoup(html, 'html.parser')
    href_list = []
    for a_elem in soup.select("h1.css-1uyzlnj"):
      href_list.append(a_elem.text)
    return  " ".join(href_list)

def content(adresa_clanku):

    r = requests.get(adresa_clanku)
    html = r.text

    soup = BeautifulSoup(html, 'html.parser')
    href_list = []
    for a_elem in soup.select(".container p"):
       href_list.append(a_elem.text)
    return  " ".join(href_list)
```
